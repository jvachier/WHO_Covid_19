```python
df_newindex_europe = df_newindex[df_newindex["WHO Region"].isin(['Europe'])].sort_values('Deaths - cumulative total per 100000 population')
df_newindex_americas = df_newindex[df_newindex["WHO Region"].isin(['Americas'])].sort_values('Deaths - cumulative total per 100000 population')
```

```python
df_newindex_europe['Deaths - cumulative total per 100000 population'].plot.bar(title='Europe',ylabel='Deaths - cumulative total per 100000 population')
```

```python
df_newindex_americas['Deaths - cumulative total per 100000 population'].plot.bar(title='Americas',ylabel='Deaths - cumulative total per 100000 population')
```

```python
g = seaborn.barplot(x=df_newindex_europe.index, y='Deaths - cumulative total per 100000 population', data=df_newindex_europe)
g.set_xticklabels(g.get_xticklabels(), rotation=90)
g.set_title('Europe')
```

```python
g = seaborn.barplot(x=df_newindex_americas.index, y='Deaths - cumulative total per 100000 population', data=df_newindex_americas)
g.set_xticklabels(g.get_xticklabels(), rotation=90)
g.set_title('Americas')
```

```python
world_Europe = world[world["continent"].isin(['Europe'])]
world_NAmerica = world[world["continent"].isin(['North America'])]
```

```python
world_Europe
```

```python
world_Europe.plot()
```

```python
array_eu=world_Europe['name'].tolist()
```

```python
df_newindex_europe=df_newindex_europe.rename(index={'Russian Federation': 'Russia','The United Kingdom':'United Kingdom'})
```

```python
df_europe = df_newindex_europe[df_newindex_europe.index.isin(array_eu)]
```

```python
world_Europe_new=world_Europe.merge(df_europe, left_on='name', right_on='Countries',
          suffixes=('_left', '_right'))
```

```python
world_Europe_new.plot("Deaths - cumulative total per 100000 population", cmap="Blues")
world_Europe_new.plot("Cases - cumulative total per 100000 population", cmap="Greens")
```

```python
world_Europe_new.plot("Deaths - newly reported in last 7 days per 100000 population", cmap="Reds")
world_Europe_new.plot("Cases - newly reported in last 7 days per 100000 population", cmap="PuRd")
```
